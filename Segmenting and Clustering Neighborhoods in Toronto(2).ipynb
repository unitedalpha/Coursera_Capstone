```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
toronto_df = pd.read_html(url)[0]
toronto_df.head()
```

```python
toronto_df.shape
```

```python
toronto_df2 = toronto_df[~ toronto_df['Borough'].str.contains('Not assigned')]
toronto_df2.reset_index(drop=True, inplace=True)
toronto_df2.head(10)
```

```python
toronto_df2.shape
```

```python
toronto_df2['Neighbourhood'].replace('Not assigned', toronto_df2['Borough'], inplace = True)
toronto_df2.head(10)
```

```python
toronto_df3 = toronto_df2.groupby(by=['Postcode','Borough'], as_index = False).aggregate(lambda x:','.join(x))
toronto_df3.head(20)
```

```python
toronto_df3.shape
```

```python
import numpy as np
import pandas as pd
coordinates = pd.read_csv('http://cocl.us/Geospatial_data')
coordinates.rename(columns={coordinates.columns[0]:'Postcode'}, inplace=True)
coordinates.head()
```

```python
coordinates.shape
```

```python
toronto_co = pd.merge(toronto_df3, coordinates, on = 'Postcode')
toronto_co.head(20)
```
